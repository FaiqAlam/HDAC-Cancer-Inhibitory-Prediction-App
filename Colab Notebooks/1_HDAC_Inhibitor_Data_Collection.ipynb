```python
import pandas as pd
from chembl_webresource_client.new_client import new_client
```

# **Searching for Target HDAC Inhibitors**

```python
#target search for HDAC
target = new_client.target
target_query = target.search('hdac')
targets = pd.DataFrame.from_dict(target_query)
targets
```

# **Selecting and retrieving the Bioactivaty data of CHEMBL2093865**

```python
selected_target = targets.target_chembl_id[3]
selected_target
```

```python
activity = new_client.activity
res = activity.filter(target_chembl_id = selected_target).filter(standard_type="IC50")
```

```python
df = pd.DataFrame.from_dict(res)
```

```python
df.head(3)
```

```python
df.standard_type.unique()
```

**Converting to CSV file named "bioactivity_data"**

```python
df.to_csv('bioactivity_data.csv', index=False)
```

# Removing Missing **Data**

```python
df2 = df[df.standard_value.notna()]
df2
```

# Data Pre-Processing to Active, Intermediate and **Ineffective**

**Labelling molecules as active, intermediate, and ineffective**

```python
bioactivity_class = []

for i in df2.standard_value:
  if float(i)>=10000:
    bioactivity_class.append("inactive")
  elif float(i) <= 1000:
    bioactivity_class.append("active")
  else:
    bioactivity_class.append("intermediate")
```

Also removing duplicate values here


# ***First and long method***

**Iterate the molecule_chembl_id to a list**

```python
mol_cid = []
for i in df2.molecule_chembl_id:
  mol_cid.append(i)
```

**Iterate canonical_smiles to a list**

```python
canonical_smiles = []
for i in df2.canonical_smiles:
  canonical_smiles.append(i)
```

**Iterate standard_value to a list**

```python
standard_value = []
for i in df2.standard_value:
  standard_value.append(i)
```

**Combine the 4 lists into a dataframe**

```python
data_tuples = list(zip(mol_cid, canonical_smiles, bioactivity_class, standard_value))
df3 = pd.DataFrame( data_tuples,  columns=['molecule_chembl_id', 'canonical_smiles', 'bioactivity_class', 'standard_value'])
```

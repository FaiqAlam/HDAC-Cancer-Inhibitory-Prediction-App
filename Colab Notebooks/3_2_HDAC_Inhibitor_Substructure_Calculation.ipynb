# Getting the XML files to calculate all the Fingerprints and Descriptors

```python
! wget https://github.com/dataprofessor/padel/raw/main/fingerprints_xml.zip
! unzip fingerprints_xml.zip
```

# Listing and Sorting the all the XML files of the Fingerprints

```python
import glob
xml_files=glob.glob("*.xml")
xml_files.sort()
xml_files
```

We are making a dictionary so that we can easily acces the above xml files using the shorter names in the dictionary

```python
FP_list = ['AtomPairs2DCount',
 'AtomPairs2D',
 'EState',
 'CDKextended',
 'CDK',
 'CDKgraphonly',
 'KlekotaRothCount',
 'KlekotaRoth',
 'MACCS',
 'PubChem',
 'SubstructureCount',
 'Substructure']
```

**Creating the Dictionary**

```python
fp = dict(zip(FP_list, xml_files))
fp
```

# Importing our dataset

```python
import pandas as pd

df = pd.read_csv('/content/bioactivity_data_Part_2_with_intermediate.csv')
df
```

# Selecting Input for Padel for Initial Dataset

```python
df2 = pd.concat([df['canonical_smiles'],df['molecule_chembl_id']], axis = 1)
df2.to_csv('molecule.smi', sep = '\t', index = False, header = False)
df2
```

# Calculating Substructure

```python
from padelpy import padeldescriptor

fingerprint = 'Substructure'

fingerprint_output_file = ''.join([fingerprint,'.csv']) #Substructure.csv
fingerprint_descriptortypes = fp[fingerprint]

padeldescriptor(mol_dir='molecule.smi',
                d_file=fingerprint_output_file, #'Substructure.csv'
                #descriptortypes='SubstructureFingerprint.xml',
                descriptortypes= fingerprint_descriptortypes,
                detectaromaticity=True,
                standardizenitro=True,
                standardizetautomers=True,
                threads=2,
                removesalt=True,
                log=True,
                fingerprints=True)
```

# Displaying Molecular Fingerprints

```python
df3 = pd.read_csv(fingerprint_output_file)
df3
```

# Changing to form that can be able to predict the pIC50 values

```python
df_X = df3.drop(columns = ['Name'])
df_X
```

```python
df_Y = df['pIC50']
df_Y
```

```python
dataset = pd.concat([df_X, df_Y], axis = 1)
dataset
```

```python
dataset.to_csv('substructure_pIC50.csv',index = False)
```
